# Kathmandu AIR QUALITY INDEX ANALYSIS (AQI)

```python
#### importing frameworks
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#### Loading Data 
df = pd.read_csv(r"D:\BootCamp\Air Pollution Analysis  in KTM\AQI.csv")
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
df.info
```

```python
def categorize_aqi(aqi):
    if aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Moderate'
    elif aqi <= 150:
        return 'Unhealthy for Sensitive Groups'
    elif aqi <= 200:
        return 'Unhealthy'
    elif aqi <= 300:
        return 'Very Unhealthy'
    else:
        return 'Hazardous'

# Apply categorization
df['AQI_Level'] = df['AQI Value'].apply(categorize_aqi)

# View results
print(df[['Date', 'AQI Value', 'AQI_Level']])
```

```python
#### plots
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Date', y='AQI Value', marker='o')
plt.title("Kathmandu AQI Trend (Q1 2025)")
plt.ylabel("AQI Value")
plt.xlabel("Date")
plt.grid(True)
plt.tight_layout()
plt.show()
```

##  Pie Chart  Proportion of AQI Status Levels
#### This will help visualize how often each AQI category appears

```python
# Count of each level
level_counts = df['AQI_Level'].value_counts()

# Plot pie chart
plt.figure(figsize=(8,8))
colors = ['green', 'yellow', 'orange', 'red', 'purple', 'maroon']
plt.pie(level_counts, labels=level_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title("Proportion of AQI Categories (Kathmandu Q1 2025)")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

## Bar Plot
#### Count of days by AQI Status

```python
#Shows how many days were unhealthy, Harzardous
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Status', palette='Reds')
plt.title("Number of Days by AQI Status")
plt.xlabel("Air Quality Status")
plt.ylabel("Number of Days")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

## Box plot : AQI distribution by days

```python

df['Month'] = df['Date'].dt.strftime('%b')
plt.figure(figsize=(20,15))
sns.boxplot(data=df, x='Date', y= 'AQI Value', palette='Reds')
plt.title("Number of Days by AQI Status")
plt.xlabel("Date")
plt.ylabel("AQI value")
plt.tight_layout()
plt.show()
```

## KPI Values  Min, Max, Average AQI
### Display key metrics.

## Highlight Worst AQI Day

```python
kpi_data = {
    'Metric': ['Max AQI', 'Min AQI', 'Average AQI'],
    'Value': [df['AQI Value'].max(), df['AQI Value'].min(), df['AQI Value'].mean()]
}
kpi_df = pd.DataFrame(kpi_data)

plt.figure(figsize=(7,5))
sns.barplot(data=kpi_df, x='Metric', y='Value', palette='coolwarm')
plt.title("AQI KPI Metrics")
plt.tight_layout()
plt.show()
```

```python
worst_day = df[df['AQI Value'] == df['AQI Value'].max()]
print(" Worst AQI Day:\n", worst_day)
```
